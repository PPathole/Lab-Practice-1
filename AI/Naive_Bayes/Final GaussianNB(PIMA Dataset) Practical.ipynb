```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C://Users/Admin/Downloads/diabetes.csv")
```

```python
dataset
```

```python
print(dataset.isnull().any())
```

```python
X = dataset.iloc[:, :-1]
X
```

```python
y = dataset.iloc[:, -1]
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
X_train
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
X_test
```

```python
X_test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
model = clf.fit(X_train, y_train)
```

```python
model
```

```python
dataset.keys()
```

```python
X_new = [[12, 80, 65, 40, 0, 30, 0.500, 45]]
print(model.predict(X_new))
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print("The accuracy of the model is: "+str(accuracy_score(y_test, y_pred)))
```

## Training set accuracy 

```python
print(model.score(X_train, y_train))
```

## Testing set accuracy

```python
print(model.score(X_test, y_test))
```

```python
print("The precision of this model is: "+str(precision_score(y_test, y_pred)))
```

```python
print("The recall of this model is: "+str(recall_score(y_test, y_pred)))
```

```python
print("The F1 score of this model is: "+str(f1_score(y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
