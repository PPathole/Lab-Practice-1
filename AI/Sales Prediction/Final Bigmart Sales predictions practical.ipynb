```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C://Users/Admin/Downloads/Train.csv")
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:,1]
```

```python
dataset.isnull().any()
```

```python
dataset.isnull().sum()
```

```python
from sklearn_pandas import CategoricalImputer
```

```python
cimputer = CategoricalImputer()
```

```python
data = np.array(X.iloc[:,8], dtype='object')

X.iloc[:,8] = cimputer.fit_transform(data)
dataset['Outlet_Size'] = X.iloc[:,8]
```

```python
dataset.iloc[:,8].isnull().any()
```

```python
dataset.isnull().any()
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values='NaN', strategy='mean')

data2 = np.array(X.iloc[:,1:2], dtype='object')

X.iloc[:,1:2] = imputer.fit_transform(data2)

dataset['Item_Weight'] = X.iloc[:,1:2]
```

```python
dataset.isnull().any()
```

```python
dataset['Item_Fat_Content'] = dataset['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()

mod = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

dataset['Item_Identifier'] = labelEncoder.fit_transform(dataset['Item_Identifier'])
dataset['Outlet_Identifier'] = labelEncoder.fit_transform(dataset['Outlet_Identifier'])

for i in mod:
    dataset[i] = labelEncoder.fit_transform(dataset[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
model = linreg.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
```

```python
print(model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
```

```python
X_test = pd.DataFrame(X_test)
y_pred = pd.DataFrame(y_pred)
```

```python
X_df = pd.DataFrame(np.array(X_test).T)
y_df = pd.DataFrame(np.array(y_pred).T)
headers = pd.DataFrame(dataset.iloc[:])

result = pd.concat([headers, X_test, y_pred], axis=1, ignore_index=True)
result.to_csv("C://Users/Admin/Documents/results.csv")
print("Check the documents to see the results")
```
